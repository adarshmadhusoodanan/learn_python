### 3. WAP TO CHECK the given number is strong or not

- Algorithm for Strong Number
- A strong number is a number whose sum of factorials of digits equals the number itself

Steps:
1. Take a number as input
2. Extract each digit from the number
3. Find the factorial of each digit
4. Add all the factorials together
5. Compare the sum with the original number
6. If sum equals the original number, it's a strong number
7. Otherwise, it's not a strong number

- Example: 145
-  1! + 4! + 5! = 1 + 24 + 120 = 145 (Strong number)
- Example: 134
-  1! + 3! + 4! = 1 + 6 + 24 = 31 ≠ 134 (Not a strong number)

```python
def factorial(n):
    fact =1
    for i in range(1,n+1):
        fact = fact*i
        # print(f"{fact} x {i}")
    return fact


num = int(input(" enter number "))
temp = num
res=0
while num!=0:
    rem = num%10
    res += factorial(rem)
    num = num//10
if res == temp:
    print("strong number")
else:
    print("not strong number")
```


### Algorithm for Perfect Number

A perfect number is a positive integer that is equal to the sum of its proper divisors (divisors excluding the number itself).

**Steps:**
1. Take a number as input
2. Find all divisors of the number (excluding the number itself)
3. Calculate the sum of all divisors
4. Compare the sum with the original number
5. If sum equals the original number, it's a perfect number
6. Otherwise, it's not a perfect number

**Examples:**
- 6: Divisors are 1, 2, 3 → 1 + 2 + 3 = 6 (Perfect number)
- 28: Divisors are 1, 2, 4, 7, 14 → 1 + 2 + 4 + 7 + 14 = 28 (Perfect number)
- 10: Divisors are 1, 2, 5 → 1 + 2 + 5 = 8 ≠ 10 (Not a perfect number)

```python
n = int(input(" enter number of terms "))
res = 0 
for i in range(1,n):
    if n%i ==0:
        res += i

if res == n:
    print("perfect number")
else:
    print("not perfect number")
```

### Algorithm for Sum of Digits

**Steps:**
1. Take a number as input
2. Extract the last digit using modulo operator (%)
3. Add the digit to a sum variable
4. Remove the last digit by integer division (//)
5. Repeat steps 2-4 until the number becomes 0
6. Print the final sum

**Example:**
- 123: 3 + 2 + 1 = 6
- 456: 4 + 5 + 6 = 15


```python
n = int(input(" enter number"))
sum =0
while n>0:
    digit = n % 10
    sum = sum + digit
    n = n // 10
print("Sum of digits:", sum)
```

### wa program to check given number is a dusarium number or not

A Disarium number is a number where the sum of its digits powered with their respective positions is equal to the original number.

**Algorithm for Disarium Number:**

1. Take a number as input
2. Extract each digit from the number one by one
3. Raise each digit to the power of its position (1-indexed)
4. Add all the powered digits together
5. Compare the sum with the original number
6. If sum equals the original number, it's a Disarium number
7. Otherwise, it's not a Disarium number

**Examples:**
- 135: 1¹ + 3² + 5³ = 1 + 9 + 125 = 135 (Disarium number)
- 175: 1¹ + 7² + 5³ = 1 + 49 + 125 = 175 (Disarium number)
- 124: 1¹ + 2² + 4³ = 1 + 4 + 64 = 69 ≠ 124 (Not a Disarium number)

```python
n = int(input(" enter number"))
l = len(str(n))
temp = n
sum = 0
while n>0:
    digit  = n % 10
    sum += digit ** l
    l -= 1
    n = n // 10

if temp == sum:
    print("disarium number")
else:
    print("not disarium number")
```

### wap to find the frequency of the number

**Steps:**
1. Take a number as input
2. Initialize a frequency dictionary
3. Extract each digit from the number
4. Update the frequency count for each digit in the dictionary
5. Print the frequency of each digit


```python
n = int(input(" enter number"))

d = {}
while n>0:
    digit = n%10
    if digit not in d:
        d[digit] = 1
    else:
        d[digit] += 1
    n = n//10

for key, value in d.items():
    print(f"Digit {key} has frequency {value}")
```

### wa program to check automorphic number or not

**Steps:**
1. Take a number as input
2. Calculate the square of the number   
3. Check if the square ends with the same digits as the original number
4. If it does, it's an automorphic number
5. Otherwise, it's not an automorphic number

**Examples:**
- 5: 5² = 25 (Automorphic number)
- 6: 6² = 36 (Automorphic number)
- 7: 7² = 49 (Not an automorphic number)
- 25: 25² = 625 (Automorphic number)

```python
n = int(input(" enter number"))
square = n * n
temp = n
l = 0
while temp > 0:
    l += 1
    temp //= 10

divisor = 1
for i in range(l):
    divisor *= 10

last_digits = square % divisor
if last_digits == n:
    print("automorphic number")
else:
    print("not automorphic number")
```

### WAP to shift all odd number to the first place and even numbers to the last place in a given number

**Steps:**
1. Take a number as input
2. Extract each digit from the number
3. Separate odd and even digits
4. Combine odd digits followed by even digits
5. Print the rearranged number
- Example:
- Input: 123456
- Output: 135246




```python

n = int(input(" enter number: "))
if n == 0:
    print("Number after rearranging:", 0)
else:
    temp = n
    odds_rev = evens_rev = 0
    odds_count = evens_count = 0

    while temp > 0:
        d = temp % 10
        if d % 2 != 0:
            odds_rev = odds_rev * 10 + d
            odds_count += 1
        else:
            evens_rev = evens_rev * 10 + d
            evens_count += 1
        temp //= 10

    def reverse_num(x):
        r = 0
        while x > 0:
            r = r * 10 + (x % 10)
            x //= 10
        return r

    odds = reverse_num(odds_rev)
    evens = reverse_num(evens_rev)


    result = odds * 10 ** evens_count + evens
    print("Number after rearranging:", result)

```

### wa program to check given number is a smith number

**Steps:**
1. Take a number as input
2. Find the prime factors of the number
3. Calculate the sum of the digits of the prime factors
4. Calculate the sum of the digits of the original number
5. Compare the two sums
- If both sums are equal, it's a Smith number

- Example:
- 4: Prime factors are 2, 2 → Sum of digits = 2 + 2 = 4 (Smith number)
- smith number examples
- 6: Prime factors are 2, 3 → Sum of digits = 2
- 22: Prime factors are 2, 11 → Sum of digits = 2 + (1 + 1) = 4 ≠ 2 + 2 = 4 (Not a Smith number)
- 27: Prime factors are 3, 3, 3 → Sum of digits = 3 + 3 + 3 = 9 ≠ 2 + 7 = 9 (Not a Smith number)
- 58: Prime factors are 2, 29 → Sum of digits = 2 + (2 + 9) = 13 ≠ 5 + 8 = 13 (Not a Smith number)



```python


def prime_factor_sum(n):
    factor = 2
    sum = 0
    print("Prime factors are: ", end="")
    while n != 1:
        if n % factor == 0:
            sum = sum + factor
            n = n // factor
            print(factor, end=" ")
        else:
            factor +=1
        print(sum)
    return sum
sum2 = 0 
while n>0:
    digit = n%10
    sum2 += digit
    n //= 10
print(sum2)

n = int(input(" enter number"))

if sum2 == prime_factor_sum(n):
    print("smith number")
else:
    print("not smith number")

```

trimorphic number
- it is a number whose cube ends with the same digits as the original number

```python
def reverse_num(x):
    r = 0
    while x > 0:
        r = r * 10 + (x % 10)
        x //= 10
    return r

n = int(input(" enter number"))

p = True

num_rev = 0
qube = n **3

while qube>0:
    digit = qube % 10
    num_rev = num_rev * 10 + digit
    qube = qube //10

    if n == reverse_num(num_rev):
        p = True
        break
    else:
        p = False

if p:
    print("trimorphic number")
else:
    print("not trimorphic number")
```

maximum subarray sum

```python
def max_subarray_sum(arr):
    max_sum = 0
    if len(arr) == 1:
        return arr[0]
    else:
        size = int(input(" enter size of array "))
        for i in range(len(arr)):
            curr_sum = 0
            if i + size >len(arr):
                return max_sum
            else:
                for j in range(i, i + size):
                    curr_sum += arr[j]
            if curr_sum > max_sum:
                max_sum = curr_sum
        return max_sum


l1 = [-2]
print("Maximum subarray sum is:", max_subarray_sum(l1))
```
